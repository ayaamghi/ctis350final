```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download('punkt_tab')
```

```python
# Define a function to clean and extract relevant text from a tweet
def clean_tweet(tweet):
    # Remove URLs (http/https) and www URLs
    tweet = re.sub(r"http\S+|www\S+", "", tweet)
    # Remove mentions (e.g., @username)
    tweet = re.sub(r"@\w+", "", tweet)
    # Optionally, remove hashtags, or extract them if relevant. Here we remove the "#" symbol.
    tweet = re.sub(r"#", "", tweet)

    # Tokenize the tweet into words
    tokens = word_tokenize(tweet)

    # Convert to lowercase
    tokens = [word.lower() for word in tokens]

    # Remove punctuation and non-alphabetic tokens, and filter out stopwords
    stop_words = set(stopwords.words("english"))
    cleaned_tokens = [
        word for word in tokens
        if word.isalpha() and word not in stop_words
    ]

    return cleaned_tokens
```

```python
from nltk import RegexpParser
```

```python
grammar = r"NP: {<DT>?<JJ>*<NN.*>+}"
chunk_parser = RegexpParser(grammar)

def extract_noun_phrases(pos_tags):
    tree = chunk_parser.parse(pos_tags)
    phrases = []
    for subtree in tree.subtrees(filter=lambda t: t.label() == "NP"):
        phrase = " ".join(word for word, pos in subtree.leaves())
        phrases.append(phrase)
    return phrases
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')  # a lightweight, fast model
```

```python
tweet_texts = russian_df_text_subset10k["content"].tolist()
corpus_embeddings = model.encode(tweet_texts, convert_to_tensor=True)

query = "There was a police incident in downtown New York"
query_embedding = model.encode(query, convert_to_tensor=True)
```

```python
# Compute cosine similarities between the query and all corpus tweets
cosine_scores = util.cos_sim(query_embedding, corpus_embeddings)[0]

# Get the top 5 most similar tweets
top_results = cosine_scores.topk(5)
for score, idx in zip(top_results[0], top_results[1]):
    print(f"Score: {score:.3f} - {tweet_texts[idx]}")
```
