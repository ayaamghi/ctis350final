```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_df_list_from_directory(path_to_directory):
    entries = os.listdir(path_to_directory)
    df_list = []
    for entry in entries:
        if entry.endswith(".csv"):
            df_list.append(pd.read_csv(path_to_directory + "/" + entry))
    return df_list
```
